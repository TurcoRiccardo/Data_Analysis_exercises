# Tasso di occupazione in Italia

Fonti: https://dati-congiuntura.istat.it/Index.aspx?DataSetCode=DCCV_TAXOCCUMENS1

```python
datiOccupazione = pd.read_csv('data/DCCV_TAXOCCUMENS1.csv')

datiOccupazione['anno'] = pd.to_datetime(datiOccupazione['TIME']).dt.year
datiOccupazione.sort_values(['anno'], ascending=True)
intervalloDiTempo = datiOccupazione.loc[(datiOccupazione['anno'] >= 2006) & (datiOccupazione['anno'] < 2025)] #19 anni


datiFiltrati_T_Y15_89 = intervalloDiTempo.loc[(datiOccupazione['Sesso'] == 'totale') & (datiOccupazione['ETA1'] == 'Y15-89') & (datiOccupazione['T_BIS'] == '2025M3G4') & (datiOccupazione['CORREZ'] == 'N'), ['anno', 'Value']]
datiFiltrati_T_Y15_89_ragguppati_anno = datiFiltrati_T_Y15_89.groupby('anno', as_index=False).mean()

datiFiltrati_M_Y15_64 = intervalloDiTempo.loc[(datiOccupazione['Sesso'] == 'maschi') & (datiOccupazione['ETA1'] == 'Y15-64') & (datiOccupazione['T_BIS'] == '2025M3G4') & (datiOccupazione['CORREZ'] == 'N'), ['anno', 'Value']]
datiFiltrati_M_Y15_64_ragguppati_anno = datiFiltrati_M_Y15_64.groupby('anno', as_index=False).mean()

datiFiltrati_F_Y15_64 = intervalloDiTempo.loc[(datiOccupazione['Sesso'] == 'femmine') & (datiOccupazione['ETA1'] == 'Y15-64') & (datiOccupazione['T_BIS'] == '2025M3G4') & (datiOccupazione['CORREZ'] == 'N'), ['anno', 'Value']]
datiFiltrati_F_Y15_64_ragguppati_anno = datiFiltrati_F_Y15_64.groupby('anno', as_index=False).mean()

datiFiltrati_T_Y25_34 = intervalloDiTempo.loc[(datiOccupazione['Sesso'] == 'totale') & (datiOccupazione['ETA1'] == 'Y25-34') & (datiOccupazione['T_BIS'] == '2025M3G4') & (datiOccupazione['CORREZ'] == 'N'), ['anno', 'Value']]
datiFiltrati_T_Y25_34_ragguppati_anno = datiFiltrati_T_Y25_34.groupby('anno', as_index=False).mean()
```

### Percentuale di Occupazione (tutte le categorie)

```python
#grafico a linee.
#ax = datiFiltrati_T_Y15_89_ragguppati_anno.plot(kind='line', x='anno', y='Value', figsize=(14,6), color='blue')
#ax.axhline(datiFiltrati_T_Y15_89_ragguppati_anno['Value'].mean(), color='red', linestyle='--', label='Media')

#grafico a box.
ax = datiFiltrati_T_Y15_89.boxplot(by='anno', figsize=(14,6), color='blue')
ax.axhline(datiFiltrati_T_Y15_89['Value'].mean(), color='red', linestyle='--', label='Media')

ax.set_title('Percentuale di Occupazione (tutte le categorie)')
ax.set_xlabel('Anni')
ax.set_ylabel('Percentuale Occupazione')
plt.suptitle('')
ax.legend()
```

# Inflazione media in Italia

Fonti: https://www.rivaluta.it/serie-inflazione-media.asp

```python
datiInflazione = pd.read_csv('data/inflazioneMediaItalia.csv')


datiInflazione.sort_values(['category'], ascending=True)
datiInflazione_filtrati = datiInflazione.loc[(datiInflazione['category'] >= 2006) & (datiInflazione['category'] < 2025)] #19 anni
media_inflazione = datiInflazione_filtrati['Inflazione media Italia'].mean()


colors = ['red' if x < 0 else 'blue' for x in datiInflazione_filtrati['Inflazione media Italia']]
ax = datiInflazione_filtrati.plot(kind='bar', x='category', y='Inflazione media Italia', figsize=(14,6), color=colors, edgecolor='black')
ax.set_title("Inflazione Annua Media Italia")
ax.set_xlabel("Anno")
ax.set_ylabel("Inflazione (%)")
ax.axhline(media_inflazione, color='green', linestyle='--', label=f"Media: {media_inflazione:.2f}%")
for i, val in enumerate(datiInflazione_filtrati['Inflazione media Italia']):
    ax.text(i, val, f"{val:.2f}%", ha='center', va='bottom' if val >=0 else 'top', fontsize=9)
ax.legend()
plt.show()
```

# Aumento prezzi in Italia (base 2015)

Fonti: https://esploradati.istat.it/databrowser/#/it/dw/categories/IT1,Z0400PRI,1.0/PRI_CONWHONAT/PRI_CONWHONAT_BRI/DCSP_NICDUE/IT1,144_125_DF_DCSP_NICDUE_2,1.0

```python
datiPrezziBase2015 = pd.read_csv('data/prezzi_annuale_Nic_base2015.csv', encoding='UTF-8', on_bad_lines='skip')
datiPrezziBase2015.sort_values(['TIME_PERIOD'], ascending=True)
datiPrezziBase2015['valore'] = pd.to_numeric(datiPrezziBase2015['Osservazione'], errors='coerce')


datiPrezziBase_generale = datiPrezziBase2015.loc[(datiPrezziBase2015['E_COICOP_REV_ISTAT'] == '00')]

datiPrezziBase_alimenti = datiPrezziBase2015.loc[(datiPrezziBase2015['E_COICOP_REV_ISTAT'] == '01')]

datiPrezziBase_abbigliamento = datiPrezziBase2015.loc[(datiPrezziBase2015['E_COICOP_REV_ISTAT'] == '03')]

datiPrezziBase_salute = datiPrezziBase2015.loc[(datiPrezziBase2015['E_COICOP_REV_ISTAT'] == '06')]
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 8))

# Primo grafico
axes[0, 0].plot(datiPrezziBase_generale['TIME_PERIOD'], datiPrezziBase_generale['valore'], color='blue')
axes[0, 0].axhline(100, color='red', linestyle='--', label='Valore 2015')
axes[0, 0].set_title('Indice generale prezzi in Italia (base 2015)')
axes[0, 0].set_xlabel('Anni')
axes[0, 0].set_ylabel('Aumento prezzi (base 100)')
axes[0, 0].grid(True)
axes[0, 0].legend()
# Secondo grafico
axes[0, 1].plot(datiPrezziBase_alimenti['TIME_PERIOD'], datiPrezziBase_alimenti['valore'], color='blue')
axes[0, 1].axhline(100, color='red', linestyle='--', label='Valore 2015')
axes[0, 1].set_title('Prezzi prodotti alimentari in Italia (base 2015)')
axes[0, 1].set_xlabel('Anni')
axes[0, 1].set_ylabel('Aumento prezzi (base 100)')
axes[0, 1].grid(True)
axes[0, 1].legend()
# Terzo grafico
axes[1, 0].plot(datiPrezziBase_abbigliamento['TIME_PERIOD'], datiPrezziBase_abbigliamento['valore'], color='blue')
axes[1, 0].axhline(100, color='red', linestyle='--', label='Valore 2015')
axes[1, 0].set_title('Prezzi abbigliamento in Italia (base 2015)')
axes[1, 0].set_xlabel('Anni')
axes[1, 0].set_ylabel('Aumento prezzi (base 100)')
axes[1, 0].grid(True)
axes[1, 0].legend()
# Quarto grafico
axes[1, 1].plot(datiPrezziBase_salute['TIME_PERIOD'], datiPrezziBase_salute['valore'], color='blue')
axes[1, 1].axhline(100, color='red', linestyle='--', label='Valore 2015')
axes[1, 1].set_title('Prezzi servizi sanitari in Italia (base 2015)')
axes[1, 1].set_xlabel('Anni')
axes[1, 1].set_ylabel('Aumento prezzi (base 100)')
axes[1, 1].grid(True)
axes[1, 1].legend()

plt.tight_layout()
plt.suptitle('')
plt.show()
```

# Popolazione in Italia 2023

Fonti: Dati estratti il 27 dic 2023 12:58 UTC (GMT) da I.Stat

```python
etaPopolazione = pd.read_csv('data/PopolazioneItaliana2023.csv', encoding='UTF-8', on_bad_lines='skip')
etaPopolazione.sort_values(['Eta'], ascending=True)

print("Popolazione totale Italiana: " + str(etaPopolazione['totale'].sum()))
print("Di cui maschi: " + str(etaPopolazione['maschi'].sum()))
print("Di cui femmine: " + str(etaPopolazione['femmine'].sum()))


fig = plt.figure(figsize=(14, 8))
gs = fig.add_gridspec(2, 3)
ax1 = fig.add_subplot(gs[0, 0])
ax2 = fig.add_subplot(gs[0, 1])
ax3 = fig.add_subplot(gs[0, 2])
ax4 = fig.add_subplot(gs[1, :])

# Primo grafico
ax1.plot(etaPopolazione['Eta'], etaPopolazione['maschi'], color='blue')
ax1.set_title('Popolazione maschile')
ax1.set_xlabel('Eta')
ax1.set_ylabel('Numero')
ax1.grid(True)
# Secondo grafico
ax2.plot(etaPopolazione['Eta'], etaPopolazione['femmine'], color='red')
ax2.set_title('Popolazione femminile')
ax2.set_xlabel('Eta')
ax2.set_ylabel('Numero')
ax2.grid(True)
# Terzo grafico
ax3.plot(etaPopolazione['Eta'], etaPopolazione['totale'], color='green')
ax3.set_title('Popolazione totale')
ax3.set_xlabel('Eta')
ax3.set_ylabel('Numero')
ax3.grid(True)
# Quarto grafico
ax4.plot(etaPopolazione['Eta'], etaPopolazione['maschi'], color='blue', label='maschi')
ax4.plot(etaPopolazione['Eta'], etaPopolazione['femmine'], color='red', label='femmine')
ax4.plot(etaPopolazione['Eta'], etaPopolazione['totale'], color='green', label='totale')
ax4.set_title('Popolazione Italiana per eta')
ax4.set_xlabel('Eta')
ax4.set_ylabel('Numero')
ax4.legend()
ax4.grid(True)


plt.tight_layout()
plt.suptitle('')
plt.show()
```
